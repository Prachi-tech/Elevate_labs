```python
## import  required liabaries 

import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px

%matplotlib inline
sns.set(style = "whitegrid")
```

```python
df = pd.read_csv('C:\\Users\\prach\\Downloads\\Titanic-Dataset.csv')
# read the given csv file
```

```python
print(df.info())
```

```python
print(df.describe())
```

```python
print(df.isnull().sum())
```

```python
df.hist(bins=30, figsize=(15,10), color = 'skyblue')
```

```python
plt.suptitle(' Histogram of numeric feature')
plt.show()
```

```python
 for col in df.select_dtypes(include=np.number).columns:
     plt.figure(figsize=(6,3))
     sns.boxplot(data=df, x=col)
     plt.title(f'Boxplot of {col}')
     plt.show()
```

```python
corr = df.select_dtypes(include=['number']).corr()
plt.figure(figsize=(12,8))
sns.heatmap(corr, annot= True, cmap="coolwarm", fmt=".2f")
plt.title("correlation matrix")
plt.show()
```

```python
sns.pairplot(df)
plt.suptitle("Pairwise relation ship", y= 1.02)
plt.show()
```

```python
fig = px.scatter_matrix(df, dimensions=df.select_dtypes(include=np.number).columns)
fig.update_layout(title="scatter matrix (potly)", width=900,height=700)
fig.show()
```
